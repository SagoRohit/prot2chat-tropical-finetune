## Cell 8 — Fix hardcoded paths in preprocess.py

```python
import os

correct_mpnn_path = os.path.abspath("ProteinMPNN")
print("Fixing paths to:", correct_mpnn_path)

with open("preprocess.py", "r") as f:
    lines = f.readlines()

new_lines = []
dir_path_written = False
i = 0
while i < len(lines):
    line = lines[i]
    if line.strip().startswith("# model_") and "load_model" in line:
        fixed = line.strip().lstrip("# ")
        fixed = fixed.replace("'ProteinMPNN/", f"'{correct_mpnn_path}/")
        fixed = fixed.replace('"/data/zcwang', f'"{correct_mpnn_path}')
        fixed = fixed.replace("'/data/zcwang", f"'{correct_mpnn_path}")
        new_lines.append(fixed + "\n")
    elif "dir_path" in line and "=" in line:
        if not dir_path_written:
            new_lines.append(f"dir_path = '{correct_mpnn_path}/'\n")
            dir_path_written = True
    elif "/data/zcwang" in line:
        line = line.replace("/data/zcwang/Protein/project/ProteinMPNN/", correct_mpnn_path + "/")
        line = line.replace("/data/zcwang/Protein/project/ProteinMPNN", correct_mpnn_path)
        new_lines.append(line)
    else:
        new_lines.append(line)
    i += 1

with open("preprocess.py", "w") as f:
    f.writelines(new_lines)

remaining = [l.strip() for l in new_lines if "/data/zcwang" in l]
commented = [l.strip() for l in new_lines if l.strip().startswith("# model_") and "load_model" in l]
print("✅ preprocess.py fixed")
print("✅ No hardcoded paths" if not remaining else f"❌ Still has: {remaining}")
print("✅ All model lines active" if not commented else f"❌ Still commented: {len(commented)}")
```

## Cell 9 — Patch demo.py (all fixes in one shot)

```python
import os, re

with open("demo.py", "r") as f:
    src = f.read()

correct_mpnn_path = os.path.abspath("ProteinMPNN")

# Fix 1: hardcoded ProteinMPNN paths
src = src.replace("/data/zcwang/Protein/project/ProteinMPNN/", correct_mpnn_path + "/")
src = src.replace("/data/zcwang/Protein/project/ProteinMPNN", correct_mpnn_path)

# Fix 2: Flask host 0.0.0.0 for ngrok
src = src.replace(
    "app.run(host='localhost', port=args.port, debug=False)",
    "app.run(host='0.0.0.0', port=args.port, debug=False)"
)

# Fix 3: float16 to fit in GPU memory
src = src.replace("torch_dtype=torch.float32", "torch_dtype=torch.float16")

# Fix 4: dual GPU memory limits
src = src.replace(
    'device_map="auto"',
    'device_map="auto", max_memory={0: "12GiB", 1: "12GiB"}'
)

# Fix 5: low_cpu_mem_usage required with device_map
src = src.replace(
    '''    model = AutoModelForCausalLM.from_pretrained(
        model_path,
        torch_dtype=torch.float16,
        device_map="auto", max_memory={0: "12GiB", 1: "12GiB"}
    )''',
    '''    model = AutoModelForCausalLM.from_pretrained(
        model_path,
        torch_dtype=torch.float16,
        device_map="auto", max_memory={0: "12GiB", 1: "12GiB"},
        low_cpu_mem_usage=True
    )'''
)

# Fix 6: LoRA with offload_folder
src = src.replace(
    'model = PeftModel.from_pretrained(model, model_id=lora_path, config=config)',
    'model = PeftModel.from_pretrained(model, model_id=lora_path, config=config, is_trainable=False, offload_folder="offload")'
)

# Fix 7: cast adapter to float16
src = src.replace(
    'adapter.eval()',
    'adapter.eval()\n    adapter = adapter.half()  # match float16 model'
)

# Fix 8: cast protein vectors to float16 at inference
src = src.replace(
    'protein_embedding = adapter(protein_vector, question_hidden_state)',
    'protein_embedding = adapter(protein_vector.half(), question_hidden_state.half())'
)

with open("demo.py", "w") as f:
    f.write(src)

# Create offload dir
os.makedirs("offload", exist_ok=True)

result = __import__('subprocess').run(
    ["python", "-m", "py_compile", "demo.py"],
    capture_output=True, text=True
)
print("✅ demo.py syntax valid" if result.returncode == 0 else "❌ " + result.stderr)
print("✅ Flask host 0.0.0.0"    if "0.0.0.0" in src else "❌ host not patched")
print("✅ float16"               if "float16" in src else "❌ float16 missing")
print("✅ dual GPU memory"       if "12GiB" in src else "❌ memory not set")
print("✅ low_cpu_mem_usage"     if "low_cpu_mem_usage=True" in src else "❌ missing")
print("✅ offload_folder"        if "offload_folder" in src else "❌ missing")
print("✅ adapter.half()"        if "adapter.half()" in src else "❌ missing")
print("✅ .half() at inference"  if "protein_vector.half()" in src else "❌ missing")
```
